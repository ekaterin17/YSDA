```python
import scipy.stats

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
```

## Задание 12

```python
data = pd.read_csv('winequality-white.csv', sep=';')
data.head()
```

```python
X = data.values[:, :-1]
y = data['quality'].values
```

```python
X = np.hstack((np.ones((X.shape[0], 1)), X))
```

```python
X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=53)
```

Обучим простую линейную регрессию на обучающей выборке и найдем MSE на трейне.

```python
w = np.linalg.inv((X_train.T.dot(X_train))).dot(X_train.T).dot(y_train)
```

```python
y_pred = X_test.dot(w)
print("MSE:", mean_squared_error(y_test, y_pred))
```

Рассмотрим оценку плотности с помощью гистограммы

```python
data = pd.read_csv('glass.dat', sep='\s+')['RI'].values
a, b = data.min(), data.max()
n = len(data)

data = (data - a) / (b - a)
```

```python
def hist_cv_score(p_hats, n, h):
    return 2. / ((n - 1) * h) - (n + 1)/ (h * (n - 1)) * np.sum(p_hats ** 2)
```

```python
def get_hist_cv_score(bin_count):
    h = (data.max() - data.min())/bin_count
    p_hats = np.histogram(data, bins=bin_count)[0] / n
    return hist_cv_score(p_hats, n, h)
```

```python
bins = np.arange(1, 201)
cv_scores = [get_hist_cv_score(x) for x in bins]

plt.figure(figsize=(10, 10))
plt.plot(bins, cv_scores)
plt.xlabel('Число бинов')
plt.ylabel('CV-score')
plt.grid()
plt.show()
```

```python
bin_best = bins[np.argmin(cv_scores)]
print("Лучшее число бинов:", bin_best)
```

```python
alpha = 0.05

f, bins = np.histogram(data, bins=bin_best, density=True)

c = scipy.stats.norm.ppf(1 - alpha / (2 * bin_best)) / 2 * np.sqrt(bin_best / n)

l = np.maximum(np.sqrt(f) - c, 0) ** 2
u = (np.sqrt(f) + c) ** 2

xs = bins[:-1] * (b - a) + a
f /= (b - a)
l /= (b - a)
u /= (b - a)

plt.figure(figsize=(10, 7))
plt.title('95% доверительный интервал')
plt.fill_between(xs, f, l, step="pre", color='aqua')
plt.fill_between(xs, u, f, step="pre", color='aqua')
plt.plot(xs, f, color='blue', drawstyle="steps", label='density')
plt.legend()
plt.show()
```

Рассмотрим ядерную оценку плотности (гауссовское ядро)

```python
data = pd.read_csv('glass.dat', sep='\s+')['RI'].values
```

```python
K = scipy.stats.norm(loc=0, scale=1).pdf
```

```python
def pairwise_differences(x):
    x = x.reshape(-1,1)
    return np.matrix(x.T - x)

def get_kernel_cv_score(h):
    diff = pairwise_differences(data)
    K2 = scipy.stats.norm(loc=0, scale=np.sqrt(2)).pdf
    return (np.sum(K2(diff / h)) - 2 * np.sum(K(diff / h))) / (h * n**2) + 2 * K(0) / (n * h)
```

```python
hs = np.linspace(0.05, 2., 200)
cv_scores = [get_kernel_cv_score(h) for h in hs]

plt.figure(figsize=(10, 10))
plt.plot(hs, cv_scores)
plt.xlabel('Ширина ядра')
plt.ylabel('CV-score')
plt.grid()
plt.show()
```

```python
h_best = hs[np.argmin(cv_scores)]
print("Лучшая ширина ядра:", h_best)
```

```python
@np.vectorize
def f_hat(x, h):
    return np.sum(K((x - data) / h)) / (n * h)

@np.vectorize
def se_hat(x, h):
    y = K((x - data) / h) / h
    return np.sqrt(np.std(y) / len(y))
```

```python
xs = np.linspace(a, b, 500)

alpha = 0.05
m = (b - a) / (3 * h_best)
q = scipy.stats.norm.ppf((1 + (1 - alpha)**(1 / m)) / 2)

f = f_hat(xs, h_best)

se = se_hat(xs, h_best)

l = np.maximum(f - q * se, 0)
u = f + q * se

plt.figure(figsize=(10, 7))
plt.title('95% доверительный интервал')
plt.fill_between(xs, f, l, step="pre", color='aqua')
plt.fill_between(xs, u, f, step="pre", color='aqua')
plt.plot(xs, f, color='blue', drawstyle="steps", label='density')
plt.legend()
plt.show()
```

## Задание 14

```python
data = pd.read_csv('glass.dat', sep='\s+')
data.head()
```

```python
y_data = data['RI'].values
x_data = data['Al'].values
```

Для решения задачи нужно построить оценку Надарая-Ватсона

```python
@np.vectorize
def r_hat(x, h):
    return np.sum(K((x - x_data) / h) * y_data) / np.sum(K((x - x_data) / h))

@np.vectorize
def calc_coef(x, h):
    return np.sum(K((x - x_data) / h))

@np.vectorize
def get_reg_score(h):
    left = (y_data - r_hat(x_data, h)) ** 2
    right = 1 / (1 - K(0) / calc_coef(x_data, h)) ** 2
    return np.sum(left * right)
```

```python
hs = np.linspace(0.06, 1., 200)
cv_scores = [get_reg_score(h) for h in hs]

plt.figure(figsize=(10, 10))
plt.plot(hs, cv_scores)
plt.xlabel('Ширина ядра')
plt.ylabel('CV-score')
plt.grid()
plt.show()
```

```python
h_best = hs[np.argmin(cv_scores)]
print("Лучшая ширина ядра:", h_best)
```

```python
sigma2 = np.sum((y_data[1:] - y_data[:-1])**2) / (2 * (n - 1))

@np.vectorize
def calc_weight(x, h):
    return K((x - x_data) / h) / np.sum(K((x - x_data) / h))

@np.vectorize
def se_hat(x, h):
    return np.sqrt(sigma2 * np.sum(calc_weight(x, h) ** 2))
```

```python
xs = np.linspace(np.min(x_data), np.max(x_data), 500)

alpha = 0.05
m = (b - a) / (3 * h_best)
q = scipy.stats.norm.ppf((1 + (1 - alpha)**(1 / m)) / 2)

r = r_hat(xs, h_best)

se = se_hat(xs, h_best)

l = r - q * se
u = r + q * se

plt.figure(figsize=(10, 7))
plt.title('95% доверительный интервал')
plt.fill_between(xs, r, l, color='aqua')
plt.fill_between(xs, u, r, color='aqua')
plt.plot(xs, r, color='blue', label='predict')
plt.xlabel('Al')
plt.ylabel('RI')
plt.legend()
plt.show()
```
