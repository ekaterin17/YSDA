## Задача 8

```python
from statsmodels.distributions.empirical_distribution import ECDF
```

```python
wines = pd.read_csv('winequality-red.csv', sep=';')
x = np.array(wines['pH'])
n = len(x)
alpha = 0.05
small_alpha = 0.01
```

```python
@np.vectorize
def asympt(f, alpha):
    return z(alpha / 2) * math.sqrt(f * (1 - f) / n)

def dvoretsky_eps(n, alpha):
    return math.sqrt(math.log(2 / alpha) / (2 * n))
```

```python
ecdf = ECDF(x)
plt.step(ecdf.x, ecdf.y, color = 'aqua', label='$\hat{F}$(x)', linewidth = 1.3) 
plt.fill_between(ecdf.x, 
                 np.minimum(1, ecdf.y + asympt(ecdf.y, alpha)), 
                 np.maximum(0, ecdf.y - asympt(ecdf.y, alpha)), 
                 color='b', alpha=0.2, label='asymptotic', step='mid')
plt.fill_between(ecdf.x, 
                 np.minimum(1, ecdf.y + dvoretsky_eps(n, alpha)), 
                 np.maximum(0, ecdf.y - dvoretsky_eps(n, alpha)), 
                 color='g', alpha=0.2, label='Dvoretsky', step='mid')

plt.xlabel('pH')
plt.ylabel('CDF')
plt.legend()
plt.show()
```

На основе решения __задачи 7__ построим оценку se и доверительный интервал.

Пусть $\theta = F(b) - F(a)$. Тогда для $\hat{\theta} = \hat{F}(b) - \hat{F}(a)$ выполнено:

$$\hat{se} = \sqrt{\frac{\hat{\theta}(1 - \hat{\theta})}{n}}$$

И, соответственно, доверительный интервал равен:

$$\hat{\theta} \pm z_{\alpha / 2} \cdot \hat{se}(\hat{\theta}) = \hat{\theta} \pm z_{\alpha / 2} \cdot \sqrt{\frac{\hat{\theta}(1 - \hat{\theta})}{n}}$$

```python
theta = ecdf(3.5) - ecdf(3.4)
theta
```

```python
se = math.sqrt(theta * (1 - theta) / n)
se
```

```python
print('95% confidence interval', theta  - z(alpha / 2) * se, theta  + z(alpha / 2) * se)
```

Посчитаем доверительные интервалы методом из [лекции](https://docviewer.yandex.ru/view/48928763/?*=Tb2HTwnzK5xEizjuGmIuxtLPJnV7InVybCI6InlhLXdpa2k6Ly93aWtpLWFwaS5zY2hvb2wueWFuZGV4LnJ1L3NoYWQvZ3JvdXBzLzIwMTcvc2VtZXN0ZXI0L3N0YXRpc3RpY3Npbm1hY2hpbmVsZWFybmluZy9sZWN0dXJlNC5wZGYiLCJ0aXRsZSI6ImxlY3R1cmU0LnBkZiIsInVpZCI6IjQ4OTI4NzYzIiwieXUiOiI3NDYwODE4ODgxNTQ5NjIzNDU2Iiwibm9pZnJhbWUiOmZhbHNlLCJ0cyI6MTU1Mjc0MTExMjIwNH0%3D) (слайд 11).

```python
data = pd.read_csv('WaterTreatment.csv')
```

```python
x_initial = np.array(data[data['reduced_microorganism'] == 0]['sediment'])
x_reduced = np.array(data[data['reduced_microorganism'] == 1]['sediment'])

x_0 = x_initial
x_1 = x_reduced
n_0 = len(x_0)
n_1 = len(x_1)
```

Для начала посчитаем: $\hat{\mu}_0 = \hat{\mu}_{initial}$, $\hat{\mu}_1 = \hat{\mu}_{reduced}$.
Также посчитаем $\hat{se}(\hat{\mu}_0)$, $\hat{se}(\hat{\mu}_1)$

```python
mu_0 = np.mean(x_0)
mu_1 = np.mean(x_1)

se_0 = np.std(x_0) / np.sqrt(n_0)
se_1 = np.std(x_1) / np.sqrt(n_1)

se = np.sqrt(se_0**2 + se_1**2)


theta = mu_0 - mu_1
print('Оценка стандартной ошибки оценки', se)
```

Теперь для статистики $\hat{\theta} = \hat{\mu}_0 - \hat{\mu}_1$, оценивающей разность матожиданий, доверительный c уровнем доверия $\alpha$ интервал строится как

$$\hat{\theta} \pm z_{\alpha / 2} \cdot \hat{se}(\hat{\theta})$$

где $\hat{se}(\hat{\theta}) = \sqrt{\hat{se}(\hat{\mu}_0) + \hat{se}(\hat{\mu}_1)}$

```python
print('95%-confidence interval', theta - z(alpha / 2)*se, theta + z(alpha / 2)*se)
print('99%-confidence interval', theta - z(small_alpha / 2)*se, theta + z(small_alpha / 2)*se)
```

У нас была гипотеза о том, что добавление реагента __не__ влияет на осадок (т.е. что истинные $\mu_1 = \mu_2$)

__Вывод:__ на 95% уровне доверия мы эту гипотезу отвергаем, на на 99%-ом --- отвергнуть не можем.

## Задача 10

```python
def assymetry(x):
    return 1 / n * np.sum((x - np.mean(x))**3) / np.std(x)**3
```

```python
def bootstrap_statistics(x, T, B):
    Xb = np.random.choice(x, (B, len(x)), replace=True)
    return np.array([T(x) for x in Xb])
```

```python
n = 50
y = np.random.normal(0, 1, 50)
x = np.exp(y)
alpha = 0.05
small_alpha = 0.01
B = 100000
```

```python
Ts = bootstrap_statistics(x, assymetry, B)
```

__1. Нормальный интервал__

```python
T = assymetry(x)
se_boot = np.std(Ts)
```

```python
print('95%-confidence normal interval', T - z(alpha / 2) * se_boot, T + z(alpha / 2) * se_boot)
```

__2. Центральный интервал__

```python
print('95%-confidence central interval', 2 * T - np.percentile(Ts, 100 * (1 - alpha / 2)), 2 * T - np.percentile(Ts, 100 * alpha / 2))
```

__3. На основе перцентилей__

```python
print('95%-confidence percentile-based interval', np.percentile(Ts, 100 * alpha / 2), np.percentile(Ts, 100 * (1 - alpha / 2)))
```

## Задача 11

```python
def strange_statistic(x):
    return np.e**(np.mean(x))
```

```python
@np.vectorize
def true_density(x, mean=10, std=1, n=100):
    if (x <= 0):
        return 0
    
    return math.sqrt(n / 2 / math.pi) / std / x * np.exp(-(1 / 2 / std**2) * n * (np.log(x) - mean) ** 2 )
```

```python
mu = 10
n = 100
B = 10000
alpha = 0.05
X = np.random.normal(mu, 1, n)
```

```python
mu_initial = strange_statistic(X)
mus = bootstrap_statistics(X, strange_statistic, B)

plt.hist(mus, bins=100, normed=True, label='empirical density')
lspace = np.linspace(np.min(mus), np.max(mus), 10000)
plt.plot(lspace, true_density(lspace, mu, 1), color='red', label='true density')
plt.legend()
plt.show()
```

```python
se_boot_mu = np.std(mus)
print('se_boot', np.std(mus))
```

```python
print('95%-confidence normal interval', mu_initial - z(alpha / 2) * se_boot_mu,
                                        mu_initial + z(alpha / 2) * se_boot_mu)

print('95%-confidence central interval', 2 * mu_initial - np.percentile(mus, 100 * (1 - alpha / 2)), 
                                         2 * mu_initial - np.percentile(mus, 100 * alpha / 2))

print('95%-confidence percentile-based interval', np.percentile(mus, 100 * alpha / 2), 
                                                  np.percentile(mus, 100 * (1 - alpha / 2)))
```
